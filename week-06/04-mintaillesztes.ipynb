```python
"""
Mintaillesztés. (Regex)
Ez független a nyelvtől, amolyan nyelv a nyelvben.
Mintaillesztés az olyan mint keresés egy szövegfileban. 
"""

#Alap illesztés, itt megadjuk mire szeretnénk rákeresni

import re #illesztéshez kell
def illeszt(szavak, minta):# bemenet szólista, minta az r-betű raw sztring azaz nem kezeli speciális karakterként a \-t.
                           # Megadja, hogy a mintaként megadott karakter miben van, ill. miben nincs.
    illeszkedik, nem_illeszkedik = [], []
    for szo in szavak:
        if bool(re.search(minta, szo)): #keresés eredménye logikaivá alakítva
            illeszkedik.append(szo)
        else:
            nem_illeszkedik.append(szo)
    print(f'illeszkedik {illeszkedik} ')
    print(f'Nem illeszkedik {nem_illeszkedik}')
    
illeszt(['kar', 'akar', 'karó', 'kavar', 'kör'], r'kar')
    
```

```python
#Illesztés speciális karakterekre.
#Sor eleje : ^
#Sor vége : $

"""
Minta   Szavak  Illeszkedik     Nem illeszkedik
^kar    kar     kar, karó       akar, kár
kar$    akar    kar, akar       karó, kár
^kar$   karo    kar             akar, karó, kár (Ez teljes illeszkedés)
        kár
"""

#Joker karakterek
# A joker a . bizonyos esetekben * ?, de mást jelent. A pont 1 karaktert jelent, tehát a k.r mintára a kar és a kár illeszkedik.

#Többszürüs előfordulás
"""
Legalább egyszeri előfordulás : +
Legfeljebb egyszeri előfordulás : ?
Tetszőleges számú előfordulás : *
Mindig a előtte levő egységekre vonatkozik.

Minta   Szavak  Illeszkedik     Nem illeszkedik
kár     kr      kár             kr, kar, kááár
ká?r    kar     kr, kár         kar, kááár
ká+r    kár     kár, kááár      kr, kar
ká*r    kááár   kr, kár, kááár  kar

ká?r    itt azt jelenti, hogy az á az opcionális
ká+r    itt azt jelenti, hogy min. 1 á-nak kell benne lennie
ká*r    itt az á akárhányszor előfordulhat, 0x is.
A .* az akármi akárhányszor.

"""

#Adott számú előfordulás
"""
pontosan m : {m}
legalább m : {m,}
legfeljebb n : {,n}
legalább m és legfeljebb n : {m,n}

Szavak : kár, káár, kááár, káááár

Minta       Illeszkedés
ká{2}r      káár
ká{2,}r     káár, kááár, káááár
ká{,3}r     kár, káár, kááár
ká{2,3}r    káár, kááár
ká{2,4}r    káár, kááár, káááár
"""

#Joker karakterek többszörözése
"""
Szavak : kr, kar, kár, kavar

Minta       Illeszkedés
k.?r        kr, kar, kár (egyvalami lehet közte)
k.+r        kar, kár, kavar (legalább 1x előfordul közte valami)
k.*r        kr, kar, kár, kavar 
"""

#Speciális karakter illeszkedést.
"""
Újsor jellegű karakterekre való illeszkedés

A \ lehet előhívni. Emiatt célszerű r' sztringet létrehozni.
\n  newline, soremelés
\r  Kocsi vissza
\n  Új sor (Linux)
\r\n    Új sor Win
\r  Új sor Mac

Szóközökre való illeszkedés
\t  Tab
\s  Fehér karakter
\S  Nem fehér karakter

Látható karakterekre való illeszkedés
\w  Alfanumerikus
\W  Nem alfanumerikus
\d  Szám
\D  Nem szám
\.  Pont

Elhelyezkedés
\A  String első karaktere
\Z  String utolsó karaktere
\b  Szó első vagy utolsó karaktere
\B  Szón belüli karakter

"""

#Intervallum illeszkedés
"""
[abcde]     Felsorolt intervallumra való illeszkedés (A felsoroltak közül az egyik OR)
[a-e]       Intervallumra való ill. (A felsoroltak közül az egyik OR)
[^a-e]      Megadott karaktereket nem tartalmazó intervallumra való ill.

Szavak : kar, kár, kér, kor, kör

Minta           Illeszkedés
k[aá]r          kar, kár
k[^aá]r         kér, kor, kör
k[a-z]r         kar, kor
k[a-zá-é]r      kar, kár, kér kor (több intervallum megadása)
"""

#Zárojelek A zárojel között lévőkre vonatkozik az utánnuk következő utasítás
"""
Szavak : kar, kár, kárkár, kárkárkár

Minta           Illeszkedés
^(kár)*$        kár, kárkár, kárkárkár ()
^(kár){2}$      kárkár (2x szerepelhet a minta)
^(kár){1,2}$    kár, kárkár
(kár){1,2}      kár, kárkár, kárkárkár (kár és a kárkár játszik, és az utóbbi a kárkárkár-ban 1x ott van.)
"""

#Vagy kapcsolat
"""
Jele : |

Szavak : kááár, kavar, kövér

Minta           Illeszkedés
k(ááá|ava)r     kááár, kavar

"""

# \1 (Tetszőleges szám lehet az 1 helyett)
"""
Vagy esetén ugyanaz mégegyszer.
(x|y)\1 esetén az x vagy y lehet, a második pedig az ami ez első volt (xx vagy yy)
(vw|xy)\1 esetén vwvw vagy xyxy
(x|y)\1\1\1 esetén xxxx vagy yyyy
Közte lehetnek egyéb minták is. (x|y) \1 \1 esetén x x x vagy y y y

Minta               Illeszkedik                     Nem illeszkedik
(x|y)\1             xx, yy                          xy, yx
(vw|xy)\1           vwvw, xyxy                      vwxy, yxvw
(vw|xy) \1 \1       vw vw vw, xy xy cy              xyxyxy, yx, xy
(v|w)(x|y)\1\2      vxvx, vyvy, wxwx, wywy          vxwy
"""
```
